## Import pandas and reading files

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

le = LabelEncoder()

df_test = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Day_01/test.csv")
df_train = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Day_01/train.csv")
```

```python
df_full = pd.concat([df_test, df_train])
df_full.info()
```

```python
df_full.head()
```

### Droping columns

```python
columns = ["PassengerId", "Name", "Ticket", "Fare", "Cabin", "Embarked"]
df_full = df_full.drop(columns = columns, axis = 1)
df_full.info()
```

```python
df_full.head()
```

```python
df_full["Survived"].value_counts()
```

### Filling NaN

```python
df_full["Age"] = df_full["Age"].fillna(df_full["Age"].median())
df_full["Survived"] = df_full["Survived"].fillna(df_full["Survived"].median())
df_full["Sex"] = le.fit_transform(df_full["Sex"])
```

```python
df_full.info()
```

```python
df_full["Age"] = df_full["Age"].astype("int64")
df_full["Survived"] = df_full["Survived"].astype("int64")
df_full.info()
```

### Formatting to csv

```python
df_full.to_csv("/content/drive/MyDrive/Colab Notebooks/Day_01/full_data.csv")
```

## End of Parsing

## Training

### Split data

```python
data = df_full.drop(columns = ["Survived"], axis = 1)
answer = df_full["Survived"].values
data_train, data_test =  train_test_split(data, train_size = 0.8, test_size=0.2)
answer_train, answer_test = train_test_split(answer, train_size = 0.8, test_size=0.2)
```

```python
print(data_train.shape, ' : ', answer_train.shape)
```

### K-Means algorythm

```python
neighbors = [3,6,9]
train_score = []
test_score = []
for n in neighbors:
    kn = KNeighborsClassifier(n_jobs=-1, n_neighbors=n)
    kn.fit(data_train, answer_train)
    #добавляем в тренировочный список
    train_score.append(kn.score(data_train, answer_train))
    #добавляем в тестовый список
    test_score.append(kn.score(data_test, answer_test))
```

```python
print("Train", train_score)
print("Test", test_score)
```

```python
print(classification_report(answer_test, kn.predict(data_test)))
```

### Random Forest algorithm

```python
rfc = RandomForestClassifier(n_jobs = -1)
rfc.fit(data_train, answer_train)
```

```python
rfc.score(data_train, answer_train)
```

```python
print(classification_report(answer_test, rfc.predict(data_test)))
```
